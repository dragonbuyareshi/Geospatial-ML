# Data Cleaning

```python
# Safely drop 'Search Parameters' if present
if 'Search Parameters' in data.columns:
    data = data.drop('Search Parameters', axis=1)
else:
    # Column not present; nothing to drop
    pass
```

```python
data.head()
```

```python
data = data.iloc[1:]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
# 1. Fill numeric columns (int and float) with 0
numeric_cols = data.select_dtypes(include=['float64']).columns
data[numeric_cols] = data[numeric_cols].fillna(0)
```

```python
# 2. Fill object/string columns with 'None'
object_cols = data.select_dtypes(include=['object']).columns
data[object_cols] = data[object_cols].fillna('None')
```
